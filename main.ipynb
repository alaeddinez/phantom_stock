```python
import pandas as pd


data = pd.read_csv("prev_global_20190121_R11_SR2.csv", sep=",")
```

```python
# create a fictif column
data["QTE_CHANTIER"] = 1

# calculate instant score
data["score_t"] = (data.QTE_PREV - data.QTE_REEL) / data.QTE_CHANTIER
data.fillna(0, inplace =True)

# filtrer sur un mag
data_exmag = data[data.NUM_ETT == 10 ]

#filtrer sur un sku
data_exsku = data_exmag[data_exmag.NUM_ART == 69725691]

data_exsku['cum_score'] = data_exsku['score_t'].cumsum()
```

```python
d = data_exsku[["NUM_ART","AN_SEM","QTE_REEL","QTE_PREV","score_t","cum_score"]]
```

```python
import numpy as np
```

```python
d['rate'] = (d.score_t.diff() / d.score_t) * 100
```

```python
d['cum_rate'] = d['rate'].cumsum()
```
